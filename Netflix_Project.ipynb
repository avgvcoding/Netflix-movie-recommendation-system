**If a producer wants to release some content, which month must he do so?( Month when least amount of content is added)**

```python
df_filtered = df[(df['Release Date'].dt.year >= 2008) & (df['Release Date'].dt.year <= 2020)]

df_filtered['year'] = df_filtered['Release Date'].dt.year
df_filtered['month'] = df_filtered['Release Date'].dt.strftime('%B')

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'][::-1]

df_monthly_counts = df_filtered.groupby(['year', 'month']).size().unstack().fillna(0)[month_order].T

plt.figure(figsize=(9, 8), dpi=200)
plt.pcolor(df_monthly_counts, cmap='afmhot_r', edgecolors='white', linewidths=2)
plt.xticks(np.arange(0.5, len(df_monthly_counts.columns), 1), df_monthly_counts.columns, fontsize=7)
plt.yticks(np.arange(0.5, len(df_monthly_counts.index), 1), df_monthly_counts.index, fontsize=7)

plt.title('Monthly Release Distribution (2008-2020)', fontsize=12, fontweight='bold', position=(0.20, 1.0 + 0.02))
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=8)
cbar.ax.minorticks_on()

plt.show()
```

# Recommendation System

```python
def clean_data(x):
        return str.lower(x.replace(" ", ""))
```

```python
features=['Title','Director','Actors','Country Availability','Summary','Genre']
df=df[features]
```

```python
for feature in features:
    df[feature] = df[feature].apply(clean_data)
```

```python
def create_soup(x):
    return x['Title']+ ' ' + x['Director'] + ' ' + x['Actors'] + ' ' +x['Country Availability']+' '+ x['Summary']+' '+ x['Genre']
```

```python
df['soup'] = df.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df=df.reset_index()
indices = pd.Series(df.index, index=df['Title'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

tfidf = TfidfVectorizer(stop_words='english')


#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['Summary'])
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations_new(Title, cosine_sim=cosine_sim):
    Title=Title.replace(' ','').lower()
    idx = indices[Title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return df['Title'].iloc[movie_indices]
```
